```python
import ipywidgets as widgets
from IPython.display import display
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

widgets.IntSlider()  # This is just to force initialization

```

```python
from lazypredict.Supervised import LazyClassifier
"""
# Now run the LazyClassifier code

lazy_clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models, predictions = lazy_clf.fit(X_train_processed, X_test_processed, y_train, y_test)
"""

RESULTS = {}

# Train and test the model for all datasets
approaches = {'Initial': ['real_data', 'synth_data'], 'Extra': ['real_data', 'real_pseudoreal_data', 'real_pseudoreal_synth_data']}

for approach, datasets in approaches.items():
    for dataset in datasets:
        # Load the preprocessed data
        X_train = pd.read_csv(f"./datasets/preprocessed/{approach}/{dataset}/X_train.csv")
        y_train = pd.read_csv(f"./datasets/preprocessed/{approach}/{dataset}/y_train.csv")
        X_test = pd.read_csv(f"./datasets/preprocessed/{approach}/{dataset}/X_test.csv")
        y_test = pd.read_csv(f"./datasets/preprocessed/{approach}/{dataset}/y_test.csv")
        
        lazy_clf = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None, random_state=42)
        models, predictions = lazy_clf.fit(X_train, X_test, y_train.values.ravel(), y_test.values.ravel())
        
        # Store the results in a dictionary
        RESULTS[f"{approach}_{dataset}"] = {
            'models': models,
            'predictions': predictions
        }
```

```python
for name, dataset in RESULTS.items():
    print(f"Dataset: {name}")
    print(f"Number of models: {len(dataset['models'])}")
    # Print the top 5 models sorted by f1 score
    top_models = dataset['models'].sort_values(by='F1 Score', ascending=False).head(5)
    print("Top 5 models:")
    for index, row in top_models.iterrows():

        print(f'Model: {index}, F1 Score: {row["F1 Score"]:.4f}, Accuracy: {row["Accuracy"]:.4f}')
        
    
```
